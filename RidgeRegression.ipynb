```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = 12*np.random.random((100,1))

y = 3*np.sin(x) + 0.2 * np.random.randn(100,1)

plt.scatter(x,y)
```

```
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=8, include_bias=True)
X = poly.fit_transform(x)

from sklearn.linear_model import Ridge
fig = plt.figure(figsize=(16,28))
for i in range(40):
  al = 3*i 
  ax = fig.add_subplot(8,5,i+1,title="alpha = {}".format(al))
  ri = Ridge(alpha=al, solver='cholesky')
  ri.fit(X,y)
  y_pred = ri.predict(X)
  ax.scatter(x,y, alpha=0.4, c='red')
  ax.scatter(x,y_pred)
```
