```
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```
x = 4*np.random.random((100,1))

y = 2 + 3*x - x**2 + 4*x**3 + 4 * np.random.randn(100,1)

plt.scatter(x,y)
```

```
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=10, include_bias=True)
X = poly.fit_transform(x)

from sklearn.linear_model import ElasticNet
fig = plt.figure(figsize=(16,28))
for i in range(40):
  al = 5*i
  ax = fig.add_subplot(8,5,i+1,title="alpha = {}".format(al))
  EN = ElasticNet(alpha=al, l1_ratio=0.5)
  EN.fit(X,y)
  y_pred = EN.predict(X)
  ax.scatter(x,y, alpha=0.4, c='red')
  ax.scatter(x,y_pred)
```
